```python
# Counting Sort
def countSort(l,k):
    l1=[0 for i in l]
    l2=[0 for i in range(k+1)]
    
    for i in range(0,len(l)):
        l2[l[i]]=l2[l[i]]+1
        
    for i in range(1,k+1):
        l2[i]=l2[i]+l2[i-1]
        
    for i in range(len(l)-1,-1,-1):
        tmp=l[i]
        tmp2=l2[tmp]-1
        l1[tmp2]=tmp
        l2[tmp]=l2[tmp]-1
    return l1

print(countSort([2,1,4,5,1,1],10))
```

```python
#LINKED LIST IMPLEMENTATION
class Node():
    def __init__(self,data):
        self.data=data
        self.next=None
        
class LinkedListImpl():
    def __init__(self):
        self.head=None
        self.len=0
    def __str__(self):
        s=self.head
        if s is None:
            return ""
        st=str(s.data)
        while s.next is not None:
            s=s.next
            st=st+"-->"+str(s.data)
        return st
    
    def add(self,data):
        if self.find(data) and self.len==1:
            self.remove_last()
        elif self.find(data):
            self.remove(data[0])
        n=Node(data)
        self.len+=1
        curr=self.head
        if self.head==None:
            self.head=n
        else:
            while(curr.next is not None):
                curr=curr.next
            curr.next=n
            
    def find(self,data):
        f=self.head
        fn=None
        c=0
        while (f is not None):
            if f.data[0]==data[0]:
                fn=f
                c=c+1
                break
            f=f.next
            c=c+1
        if fn is not None:
            return True
        else:
            return False
        
    def get_val(self,data):
        f=self.head
        fn=None
        c=0
        while (f is not None):
            if f.data[0]==data:
                fn=f.data
                c=c+1
                break
            f=f.next
            c=c+1
        if fn is not None:
            print(fn[1])
            #return fn
        else:
            print("Key not found",data)
            #return "Key not found",data
        
    def remove(self,data):
        r=self.head
        #print(self.len)
        while(r is not None):
            if r.data[0]==data:
                if r.next is None:
                    self.remove_last()
                    
                    #break
                else:
                    r1=r.next
                    d=r1.data
                    r.next=r.next.next
                    r.data=d
                    #break;
                self.len-=1
                break                
            else:
                r=r.next
                
    def remove_last(self):
        if self.len==1:
            self.head=None
        else:
            r=self.head
            for i in range(0,self.len):
                if i==self.len-1:
                    r.next=None
                else:
                    r=r.next
        self.len-=1
                
```

```python
# Hash Table implementation using linear chaining. 
# The key value pairs are stored in the form of tuples. 
# If same key is inserted again, its value gets replaced with new one

HashTable=[LinkedListImpl() for i in range(11)]
def Hash_fun(k):
    return k%11

def insert(k,v):
    HashTable[Hash_fun(k)].add((k,v))
    
def get_val(k):
    HashTable[Hash_fun(k)].get_val(k)
    
def delete(k):
    HashTable[Hash_fun(k)].remove(k)
    
insert(1,2)
insert(12,3)
insert(23,4)
delete(12)
insert(10,2)
insert(11,5)
insert(10,3)
delete(11)
get_val(10)
get_val(11)
for i in HashTable:
    print(i)
```
